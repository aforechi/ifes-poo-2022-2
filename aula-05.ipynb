## O comando if (se, em português)

O comando if (se) é uma estrutura de controle condicional utilizada para representar decisões.

Uma estrutura de controle condicional executa uma ou mais instruções mediante à avaliação de uma condição.

A notação do comando **if** é

if expressão: 

        #instruções a serem executadas em caso verdadeiro;

else:

        #instruções a serem executadas em caso falso;


Exemplo:

```python
# Alterne os valores da expressão lógica entre true e false e veja o resultado

expressão = False

if expressão: 
    #instruções a serem executadas em caso verdadeiro;
    print("Espressão lógica é verdadeira!")
else:
    #instruções a serem executadas em caso falso;
    print("Espressão lógica é falsa!")
```

Se a expressão lógica do comando **if** for avaliada como verdadeira (valor diferente de 0), são executadas as instruções do bloco **if**, senão, serão executadas as instruções do bloco **else**.

Se o **if** ou **else** podem ter apenas uma instrução ou uma sequencia de instruções.

A cláusula **else** é opcional, mas é fortemente recomendada para que haja um feedback do programa em relação ao seu fluxo de execução.


Exemplo: O algoritmo a seguir verifica se o usuário acertou um número mágico sorteado pelo programa. Na linguagem Python, para sortearmos números, existe a função randint() da biblioteca random.

```python
from random import randint

numero_sorteado = randint(1, 10)
```

```python
meu_palpite = int(input("Coloque o seu palpite entre 0 e 9:"))
    
if meu_palpite == numero_sorteado:
    print("Certo!!")
else:
    if meu_palpite > numero_sorteado:
        print("Palpite é maior")
    else:
        print("Palpite é menor")
```

## O comando ternário **if** : 

Você pode usar o comando **if-else** de uma forma mais compacta que permite avaliar expressões lógicas e fazer atribuições com base no resultado da expressão lógica numa única linhas.

O comando ternário **if** é chamado de ternário porque ele requer três operandos:

<instrução 1> if <expressão> else <instrução 2>

Exemplo: No exemplo abaixo, o valor 100 é atribuído a Y. Se X fosse menor que 9, o valor 200 seria atribuído a Y. Teste!

```python
X = 10
Y = 100 if X>9 else 200
print(Y)
```

A linguagem Python considera o número 0 como uma representação de FALSO e os números diferentes de 0, como uma representação de VERDADEIRO. 

Para verificar se um valor é diferente de zero, pode-se usar a seguinte avaliação através da estrutura if:

if &lt;valor&gt;:

```python
if X:
    print("Numero diferente de 0!!")
```

## O comando if aninhado

O comando if aninhado (ou ladder) é uma estrutura if inserida dentro de uma outra estrutura if.  Utilizamos o if quando temos duas opções de avaliação de uma condição. Utilizamos o if aninhado quando temos mais de duas opções.

Exemplo: informe três números e mostre o maior deles.

```python
a=100; b=20; c=30

if (a>=b) and (a>=c):
     maior = a
else:
     if b>=c:
          maior = b
     else:
          maior = c   
          
print("O maior número é ", maior)
```

## O comando elif

Quando temos o objetivo de realizar apenas comparações de igualdades e uma estrutura if aninhada se torna uma longa escada, como no exemplo a seguir, use a estrutura elif.

```python
numero = 3

if (numero==1):
     print('Opção 1')
else:
    if (numero==2):
           print('Opção 2')
    else: 
        if (numero==3):
              print('Opção 3')
        else: 
            if (numero==4):
                  print('Opção 4')
            else:
                    print('Opção default')
```

```python
numero = 3

if (numero==1):
     print('Opção 1')
elif (numero==2):
      print('Opção 2')
elif (numero==3):
      print('Opção 3')
elif (numero==4):
      print('Opção 4')
else:
      print('Opção default')
```

Exemplo: 

```python
# Informe um número de 1 a 3
CADASTRO = 1; CONSULTA = 2; EXCLUSAO = 3
```

```python
numero = int(input("Digite a operação desejada:"))
if numero == CADASTRO:
      print("Executando cadastro....")
elif numero == CONSULTA:
      print("Executando consulta....")
elif numero == EXCLUSAO:
      print("Executando exclusão de registro...")
else:
      print("Opção inválida!")
```
